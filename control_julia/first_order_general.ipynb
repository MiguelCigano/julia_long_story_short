```julia
using DifferentialEquations
using Plots

function first_order_system!(dy, y, p, t)
    τ, K, u = p
    dy[1] = (-y[1] + K * u(t)) / τ
end
```

```julia
# Parámetros del sistema
τ = 1.0      # Constante de tiempo
# u(t) = t  # Rampa lineal
u(t) = t^2   # Rampa lineal
K = 2.5      # Ganancia
# u(t) = 1.0   # Entrada constante (escalón)
```

```julia
# Condiciones iniciales
y0 = [0.0]   # Estado inicial
tspan = (0.0, 5.0) # Intervalo de simulación
p = (τ, K, u)
```

```julia
# Resolver el sistema
prob = ODEProblem(first_order_system!, y0, tspan, p);
sol = solve(prob);
```

```julia

# Graficar los resultados
plot(sol, xlabel="Tiempo", ylabel="Salida", title="Respuesta de un sistema de primer orden", legend=false)
```
